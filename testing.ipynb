```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


test_data_model_one = {
    1: [(1, 2, 'No'), (0, 2, 'Yes'), (1, 1, 'No'), (1, 1, 'No'), (1, 0, 'No')],
    2: [(1, 1, 'No'), (0, 2, 'Yes'), (1, 2, 'No'), (0, 2, 'Yes'), (1, 2, 'No')],
    3: [(1, 0, 'No'), (1, 2, 'No'), (0, 1, 'Yes'), (1, 2, 'No'), (0, 1, 'Yes')],
    4: [(0, 2, 'No'), (0, 2, 'Yes'), (1, 0, 'No'), (0, 0, 'No'), (0, 1, 'Yes')],
    5: [(1, 1, 'No'), (1, 0, 'No'), (1, 0, 'No'), (0, 0, 'Yes'), (1, 2, 'No')]
}

# Initialize a 2x3 grid for correct and incorrect guesses
correct_grid = np.zeros((2, 3), dtype=int)
incorrect_grid = np.zeros((2, 3), dtype=int)

# Iterate through test data to populate grids
for test_results in test_data_model_one.values():
    for (row, col, result) in test_results:
        if result == 'Yes':
            correct_grid[row, col] += 1
        else:
            incorrect_grid[row, col] += 1

# Plotting heatmaps
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Correct guesses heatmap
sns.heatmap(correct_grid, annot=True, fmt="d", cmap="Greens", ax=ax[0], cbar=False)
ax[0].set_title('Correct Guesses')
ax[0].set_xlabel('Column')
ax[0].set_ylabel('Row')

# Incorrect guesses heatmap
sns.heatmap(incorrect_grid, annot=True, fmt="d", cmap="Reds", ax=ax[1], cbar=False)
ax[1].set_title('Incorrect Guesses')
ax[1].set_xlabel('Column')
ax[1].set_ylabel('Row')

plt.tight_layout()
plt.show()
```
